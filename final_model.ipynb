```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('./phone_usage_cleaned.csv')
df = df.drop(['Unnamed: 0', 'User ID'], axis = 1)
df.head()
```

```python
x = df.drop('Primary Use', axis =1)
y = df['Primary Use']
```

```python
scaler = MinMaxScaler()
```

```python
x.info()
```

```python
scaler.fit(x[['Age', 'Screen Time (hrs/day)', 'Data Usage (GB/month)',
       'Calls Duration (mins/day)', 'Number of Apps Installed',
       'Social Media Time (hrs/day)', 'E-commerce Spend (INR/month)',
       'Streaming Time (hrs/day)', 'Gaming Time (hrs/day)',
       'Monthly Recharge Cost (INR)']])
```

```python
import pickle
with open('minmax_scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
x[['Age', 'Screen Time (hrs/day)', 'Data Usage (GB/month)',
       'Calls Duration (mins/day)', 'Number of Apps Installed',
       'Social Media Time (hrs/day)', 'E-commerce Spend (INR/month)',
       'Streaming Time (hrs/day)', 'Gaming Time (hrs/day)',
       'Monthly Recharge Cost (INR)']] = scaler.transform(x[['Age', 'Screen Time (hrs/day)', 'Data Usage (GB/month)',
       'Calls Duration (mins/day)', 'Number of Apps Installed',
       'Social Media Time (hrs/day)', 'E-commerce Spend (INR/month)',
       'Streaming Time (hrs/day)', 'Gaming Time (hrs/day)',
       'Monthly Recharge Cost (INR)']])
```

```python
print("Min Values:", scaler.data_min_)
print("Max Values:", scaler.data_max_)
```

```python
x = pd.get_dummies(x, dtype = int)
```

```python
x.info(0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=421, weights= 'distance' )
```

```python
model_knn.fit(x_train, y_train)
```

```python
test_score = model_knn.score(x_test, y_test)
train_score = model_knn.score(x_train, y_train)

print(f'Train Score: {train_score}')
print(f'test Score: {test_score}')
```
